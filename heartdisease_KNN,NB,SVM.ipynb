```python
# This Python 3 environment comes with many helpful analytics libraries installed
# For example, here's several helpful packages to load
#import the libraries that we needed
import numpy as np #Linear algebra
import pandas as pd #data analyse and manipulation ,eg:read_csv
#reading the dataset
df=pd.read_csv('/content/heart_missing.csv')
df
```

```python
#Displaying the first rows
df.head()
```

```python
#Displaying last few rows
df.tail()
```

```python
#Checking the datatypes
df.dtypes
```

```python
# checking the null values
df.isna().sum()
```

```python
#error correction
p=df['cp'].mean()
df['cp'].fillna(p,inplace=True)
q=df['fbs'].mean()
df['fbs'].fillna(q,inplace=True)
r=df['restecg'].mean()
df['restecg'].fillna(r,inplace=True)
s=df['oldpeak'].mean()
df['oldpeak'].fillna(s,inplace=True)
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
# Traing and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# Normalisation
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=9)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
mat=confusion_matrix(y_test,y_pred)
print(mat)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
y_pred1=nb.predict(x_test)
y_pred1
```

```python
mat=confusion_matrix(y_test,y_pred1)
print(mat)
```

```python
score=accuracy_score(y_test,y_pred1)
score
```

```python

from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
y_pred2=model.predict(x_test)
y_pred2
```

```python
y_test
```

```python
mat=confusion_matrix(y_test,y_pred2)
print(mat)
```

```python
score=accuracy_score(y_test,y_pred2)
score
```
