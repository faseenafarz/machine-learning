```python
# This Python 3 environment comes with many helpful analytics libraries installed
#Import the libraries
import numpy as np #Linear algebra
import pandas as pd # data analysis and manipulation eg:reading the dataset(read_csv)
#Reading the dataset
df=pd.read_csv('/content/Credit Score Classification Dataset (1).csv')
df
```

```python
#Checking the null values
df.isna().sum()
```

```python
g=df['Gender'].mode()[0]
g
```

```python
df['Gender'].fillna(g,inplace=True)
i=df['Income'].mean()
df['Income'].fillna(i,inplace=True)
```

```python
df.isna().sum()
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Gender']=lab.fit_transform(df['Gender'])

```

```python
df['Education']=lab.fit_transform(df['Education'])
```

```python
df['Marital Status']=lab.fit_transform(df['Marital Status'])
```

```python
df['Home Ownership']=lab.fit_transform(df['Home Ownership'])
df
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
y_pred=nb.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
mat=confusion_matrix(y_test,y_pred)
print(mat)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score
```
